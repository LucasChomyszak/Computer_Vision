```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import skimage as sk

```

```python
Barc = sk.io.imread('/content/Barcelona.jpg')
```

```python
Gray_Barc=sk.color.rgb2gray(Barc)
Gray_Barc2 = (.2125*Barc[:,:,0]+.7154*Barc[:,:,1]+.0721*Barc[:,:,2])/255

#IG_manual = (0.2125*IC[:,:,0]+0.7154*IC[:,:,1]+0.0721*IC[:,:,2])/255
```

```python
#fig=plt.figure()

fig,axes = plt.subplots(1,2,figsize=(12,7))
ax = axes.ravel() #Â convert axes to a 1d array
im1 = ax[0].imshow(Gray_Barc,cmap=plt.cm.gray)
im2= ax[1].imshow(Gray_Barc2,cmap=plt.cm.gray)

plt.show()
```
